```python
#changing the data type of disconted_price and actual_price

df['discounted_price'] = df['discounted_price'].astype(str).str.replace("‚Çπ",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].astype(str).str.replace("‚Çπ",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
#changing Datatype and values in Discount Percentage

df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%','').astype('float64')
df['discoun_percentage'] = df['discount_percentage']/ 100
```

```python
#Finding unusal string in rating column

df['rating'].value_counts()
```

```python
#check the strange row
df.query('rating == "|"')
```

```python
#changing Rating columns Datatype
df['rating'] = df['rating'].astype(str).str.replace('|', '3.9').astype('float64')
```

```python
#changing rating_count' column Data type
df['rating_count'] = df ['rating_count'].astype(str).str.replace(',','').astype('float64')
```

```python
#Impute missing values
df['rating_count'] = df.rating_count.fillna(value = df['rating_count'].median())
```
